```python
import pandas as pd
df = pd.read_csv('co2_emission.csv')
df.head()
```

Get the code of each country in a list

```python
countries = []
for i in range(len(df)):
    if df['Code'][i] in countries:
        continue
    countries.append(df['Code'][i])
```

The next two cells will check which country released the most CO2 into the Earth. The unit of measurement will be in tons

```python
total_CO2_per_country = {}
for i in range(len(df)):
    if df['Code'][i] in total_CO2_per_country:
        total_CO2_per_country[df['Code'][i]] += (df['Annual CO₂ emissions (tonnes )'][i])
        continue
    total_CO2_per_country[df['Code'][i]] = df['Annual CO₂ emissions (tonnes )'][i]
```

```python
most_CO2 = ('', 0.0)
for key in total_CO2_per_country:
    if total_CO2_per_country[key] > most_CO2[1]:
        most_CO2 = (key, total_CO2_per_country[key])
print('The country that releases the most C02 is {} ({} tons)'.format(most_CO2[0], most_CO2[1]))
```

The following cells will determine a prediction using linear regression of what the future will look like for the United States

```python
%matplotlib inline
import numpy as np
from pylab import *

# The dataframe gets split into a dataframe of data that just contains the US's data
united_states_df = df[df['Code'] == 'USA']
years = united_states_df['Year']
US_CO2 = united_states_df['Annual CO₂ emissions (tonnes )']
xlabel('Years')
ylabel('CO2')
title('History of CO2 released')
scatter(years, US_CO2)
```

```python
from scipy import stats
slope, intercept, r_value, p_value, std_err = stats.linregress(years, US_CO2)
```

The r value will be low because it will be difficult to find a straight line within in the graph

```python
r_value ** 2
```

```python
import matplotlib.pyplot as plt

def predict(x):
    return slope * x + intercept

fitline = predict(years)
plt.scatter(years, US_CO2)
plt.plot(years, fitline, c='r')
plt.show()
```

```python
val = list(years)
new_years = [i for i in range(val[0], val[-1])]
last_year = new_years[-1]
for i in range(100):
    new_years.append(last_year + i)
new_years[-1]
new_year_series = pd.Series(new_years)
new_year_series
```

```python
fitline_predict = predict(new_year_series)
plt.scatter(years, US_CO2)
plt.plot(new_year_series, fitline_predict)
plt.show()
```

```python
mean = 0
for i in range((len(fitline_predict) - 100), len(fitline_predict)):
    mean += fitline_predict[i]
mean = mean / 100
print('The average amount of CO2 per year is expected to be {}, according to this linear regression'.format(mean))
```
